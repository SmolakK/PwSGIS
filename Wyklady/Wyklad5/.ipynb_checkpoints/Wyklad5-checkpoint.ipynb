Pluginy przechowywane są w folderze własnym QGISa.
Dla QGIS3:

Zeby wtyczka działała w folderze musi znajdować się conajmniej:

    __init__.py
    plik_glowny.py
    form.py
    


<h2> __init__.py </h2>

To plik dzięki któremu wtyczka jest inicjalizowana w QGIS podczas jego uruchamiania. Jego zawrtość:

```python
# -*- coding: utf-8 -*-
def classFactory(iface):
	from plik_glowny import plik_glowny
	return plik_glowny(iface)
```

Przyjrzymy się wtyczce na przykładzie: plik_glowny nazywa się "cdr_gen", więc:

```python
# -*- coding: utf-8 -*-
def classFactory(iface):
	from cdr_gen import cdr_gen
	return cdr_gen(iface)
```

Nazwa klasy importowana jest w pliku $__init__.py$

Klasa wtyczki musi posiadać obowiązkowe funkcje:

    __init__ (jak każda klasa w Pythonie)
    initGui
    unload
    run

Funkcja $__init__$ inicjuje tylko interfejs:

```python
def __init__(self,iface):
		self.iface=iface
```

Funkcja initGui inicjuje wtyczkę w interfejsie QGISa. To jest miejsce w którym ma się pojawić, jaką ma mieć ikonkę, opis itp.

```python
	def initGui(self):
		self.action = QAction(QIcon("icon.png"), "Synthetic CDR Generator",self.iface.mainWindow())
		self.action.setObjectName("Open")
		self.action.setWhatsThis("Desc")
		self.action.setStatusTip("Tip")
		
		self.menu = QMenu(self.iface.mainWindow())
		self.menu.setObjectName("CDR")
		self.menu.setTitle("CDR")
		self.menu.addAction(self.action)
		
		self.action.triggered.connect(self.run)
		
		menuBar = self.iface.mainWindow().menuBar()
		menuBar.insertMenu(self.iface.firstRightStandardMenu().menuAction(), self.menu)
```

Funkcja unload obsługuje wydarzenie, które następuje po zamknięciu wtyczki.

```python
	def unload(self):
		self.menu.deleteLater()
```
